```python
top_12_countries = data_one["Country Name"].value_counts().head(12)

#All about Plotting
bar_width = 0.9
x = range(len(top_12_countries.index))

plt.bar(top_12_countries.index, top_12_countries.values, width=bar_width)
plt.title("Distribution_Of_Top_12_Countries")
plt.xlabel("Country_Name")
plt.ylabel("Count")

plt.xticks(x, top_12_countries.index, rotation=45)
plt.tight_layout()
plt.show()
```
