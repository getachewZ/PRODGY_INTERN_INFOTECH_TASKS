```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Handling alerts seemingly as an error but not actually an errors.
import warnings 
warnings.filterwarnings("ignore")

# Handling the plotings so as to be in the NoteBook rather than in a separated windows.
%matplotlib inline 
plt.rcParams["figure.figsize"] = 8, 3 # the width 8 and height of 4 is the out put window.
```

```python
# Twitter training and testing or validation data set definition
file_path_Train= "D:\Projects\DataScinece\PRODIGY_DS\Track_05\\AccidentsBig.csv"

# Reading Twitter training data using pandas 
AccFile = pd.read_csv(file_path_Train,delimiter  = ",")
AccFile.head()
```

```python
# Display the data frame columns of the dataset
AccFile.columns
```

```python
# Data's distribution and key metrics description
AccFile.describe()
```

```python
# Convert 'Date' to datetime with the correct format:
AccFile['Date'] = pd.to_datetime(AccFile['Date'], format='%d-%m-%Y')
```

```python
# Analyze patterns related to road conditions:
road_condition_counts = AccFile['Road_Surface_Conditions'].value_counts()
print(road_condition_counts)
```

```python
# Analyze patterns related to weather:
weather_counts = AccFile['Weather_Conditions'].value_counts()
print(weather_counts)
```

```python
# Count Accidents by Day of the Week:
weekday_accidents = AccFile.groupby(AccFile['Day_of_Week'])['Accident_Severity'].count()
print(weekday_accidents)
```

```python
# Convert 'Date' to datetime if not already done
AccFile['Date'] = pd.to_datetime(AccFile['Date'], format='%d-%m-%Y')  # Adjust format as needed

# Extract the day of the week
AccFile['Day_of_Week'] = AccFile['Date'].dt.day_name()
```

```python
# Visualize accident based on road conditions
road_condition_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
sns.countplot(data=AccFile, x='Road_Surface_Conditions', order=AccFile['Road_Surface_Conditions'].value_counts().index, palette=road_condition_colors)
plt.title('Accidents by Road Surface Condition')
plt.xlabel('Road Surface Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Visualize accidents based on weather conditions
weather_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
sns.countplot(data=AccFile, x='Weather_Conditions', order=AccFile['Weather_Conditions'].value_counts().index, palette=weather_colors)
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Converting the date in the correct date time formate
AccFile['Date'] = pd.to_datetime(AccFile['Date'], format='%d-%m-%Y')
```

```python
# Extracting Hour from the Time column
AccFile['Hour'] = pd.to_datetime(AccFile['Time'], format='%H:%M').dt.hour
hourly_accidents = AccFile.groupby('Hour')['Accident_Severity'].count()
```

```python
# Grouping accident via the days of a week
weekday_accidents = AccFile.groupby('Day_of_Week')['Accident_Severity'].count()
print(weekday_accidents)
```

```python
# Visualize accidents by day of the week:
weekday_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']
sns.barplot(x=weekday_accidents.index, y=weekday_accidents.values, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette=weekday_colors)
plt.title('Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Visualize accidents by hour of the day:
plt.fill_between(hourly_accidents.index, hourly_accidents.values, color='skyblue', alpha=0.5)
plt.plot(hourly_accidents.index, hourly_accidents.values, color='blue', alpha=0.8)
plt.title('Accidents by Hour of the Day (Area Chart)')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.xticks(range(0, 24))
plt.grid()
plt.show()
```
